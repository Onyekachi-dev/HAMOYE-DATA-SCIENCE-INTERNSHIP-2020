# Data visualization

```python
#This shows the relationship commonly between used fuel types in each plant
%matplotlib notebook
plt.pie(df.fuel_type_code_pudl.value_counts(),colors='bcgmrwy',radius=1,data=df,autopct='%.1f%%',startangle=180,labels=df.fuel_type_code_pudl.unique())
plt.title('Pie chart of commonly used fuel types by most plants')
plt.show()
```

```python
#this shows the number of plants using the different fuel types
plt.figure(figsize=(8,6))
df.fuel_type_code_pudl.value_counts().plot(kind='bar',color=["m", "g",'r','b','y','c'])
plt.xlabel('Fuel types')
plt.ylabel('Number of Plants')
plt.title('Chart showing Number of Plants against Fuel types')
plt.show()
```

```python
#This plot shows the commonly used fuel units by plants
%matplotlib notebook
sns.barplot(df.fuel_unit.value_counts().index,df.fuel_unit.value_counts())
plt.xlabel('Fuel units')
plt.ylabel('Number of Plants')
plt.title('Chart showing Number of Plants against Fuel Units')
plt.show()

```

```python
%matplotlib notebook
sns.boxplot(x="fuel_type_code_pudl", y="utility_id_ferc1",palette=["m", "g",'r','b','y','c'], data=df)
plt.xlabel('Fuel types')
plt.ylabel('utility_id_ferc1')
plt.title('Boxplot of utility_id_ferc1 against fuel types')
plt.show()
```

```python
#to show the extent of correlation between features and labels in dataset using pearson, kendall or spearman correlation
df.corr()
```

```python
#Seaborns heatmap is used to obtain the correlation between features in the dataset
%matplotlib notebook
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True)
plt.title('Heatmap of correlation between dataset features')
plt.show()

#Analysis of heatmap
#from the heatmap above, it shows that there is litlle or no correlation between the features in the dataset. This could be as a result of some outliers in the dataset.
```
